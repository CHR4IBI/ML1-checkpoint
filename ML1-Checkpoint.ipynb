```python
# Data collecting 
import pandas as pd
import numpy as np

data = pd.read_csv("kc_house_data.csv")
data.head()
```

```python
# Data preprocessing and feature selection

data["sqft_surface"] = data["sqft_living"] + data["sqft_lot"] + data["sqft_above"] + data["sqft_basement"] + data["sqft_living15"] +  data["sqft_lot15"]
data.loc[data["yr_renovated"] != 0, "seniority"] = 2021 - data["yr_renovated"]
data.loc[data["yr_renovated"] == 0, "seniority"] = 2021 - data["yr_built"]
data.drop(["id", "date", "sqft_living", "sqft_lot", "sqft_above", "sqft_basement", "sqft_living15", 'sqft_lot15', "yr_built", "yr_renovated", "zipcode"], axis = 1, inplace = True)
data.head()
```

```python
# Linear Regression // Predicting the price of a house from the grade
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

x = data["grade"].values.reshape(-1,1)
y = data["price"].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)

lg = LinearRegression()
lg.fit(x_train, y_train)
predicted = lg.predict(x_test)

fig, ax = plt.subplots(figsize=(17, 17))

ax.scatter(x,y,color="r")
plt.title("Linear Regression")
plt.ylabel("Price")
plt.xlabel("Grade")
ax.plot(x,lg.predict(x),color="k")
plt.show()

print('Linear regression')
print("MSE", mean_squared_error(y_test,predicted))
print(f"This linear model is precise at {metrics.r2_score(y_test,predicted) * 100} %")


```

```python
# Multi-linear regression // Predicting the price of a house from many features
x=data.drop("price", axis = 1)
y=data["price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)

lg = LinearRegression()
lg.fit(x_train, y_train)
predicted = lg.predict(x_test)



print("Multi-linear regression")
print("MSE", mean_squared_error(y_test,predicted))
print(f"This multi-linear model is precise at {metrics.r2_score(y_test,predicted) * 100} %")
```

```python
# Polynomial regression // Predicting the price of a house from many features
from sklearn.preprocessing import PolynomialFeatures

x=data.drop("price", axis = 1)
y=data["price"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)

poly = PolynomialFeatures(degree = 2) 
lg = LinearRegression()

x_train_fit = poly.fit_transform(x_train)
lg.fit(x_train_fit, y_train)
x_test_ = poly.fit_transform(x_test)
predicted = lg.predict(x_test_)



print("Polynomial regression")
print("MSE", mean_squared_error(y_test,predicted))
print(f"This polynomial model is precise at {metrics.r2_score(y_test,predicted) * 100} %")

```
